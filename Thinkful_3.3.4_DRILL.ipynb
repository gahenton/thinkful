```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Answer to question #4: The situations where the mean and standard deviation
# are useful are where the data looks like a bell curve, or normal distribution.
# The second example, the bernoulli coin flip set of tests, is not helped by this.

# normal, binomial, gamma, poisson, geometric, and exponential distributions
```

```python
# normal distribution
# Creating a data frame to hold the simulated ecommerce data, and populating it with a
# normally distributed variable with mean 75 and standard deviation 25.

ecommerce = pd.DataFrame()
ecommerce['spending'] = np.random.normal(75, 25, 100)

# Plot a histogram.
plt.hist(ecommerce['spending'])

# Add a vertical line at the mean.
plt.axvline(ecommerce['spending'].mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(ecommerce['spending'].mean() + ecommerce['spending'].std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(ecommerce['spending'].mean() - ecommerce['spending'].std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```

```python
# binomial distribution
# result of flipping a coin 1 time, with probability of .5, tested 100 times
# Generate a bernoulli distribution with p =0.5.
bernoulli= np.random.binomial(1, .5, 100)

#Plot a histogram.
plt.hist(bernoulli)

# Add a vertical line at the mean.
plt.axvline(bernoulli.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(bernoulli.mean() + bernoulli.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(bernoulli.mean() - bernoulli.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram
plt.show()
```

```python
# another binomial distribution
# result of flipping a coin 20 times, with probability of .5, tested 100 times.
# Generate a binomial distribution with n=20 and p=0.5.
binomial = np.random.binomial(20, 0.5, 100)

# Plot a histogram.
plt.hist(binomial)

# Add a vertical line at the mean.
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(binomial.mean() - binomial.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
# gamma distribution
# Generate a gamma distribution with shape = 5 and scale = 1
gamma = np.random.gamma(5,1, 100)

# Plot a histogram.
plt.hist(gamma)

# Add a vertical line at the mean.
plt.axvline(gamma.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(gamma.mean() + gamma.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(gamma.mean() - gamma.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
# poisson distribution
# Generate a Poisson distribution with lambda = 3
poisson = np.random.poisson(3, 100)

# Plot a histogram.
plt.hist(poisson)

# Add a vertical line at the mean.
plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(poisson.mean() + poisson.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(poisson.mean() - poisson.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
# geometric distribution
geometric = np.random.geometric(p=0.35, size=100)

# Plot a histrogram
plt.hist(geometric)

# Add a vertical line at the mean.
plt.axvline(geometric.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(geometric.mean() + geometric.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(geometric.mean() - geometric.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram
plt.show()
```

```python
# exponential distribution
exponential = np.random.exponential(scale=1.0, size=100)

# Plot a histogram
plt.hist(exponential)

# Add a vertical line at the mean.
plt.axvline(exponential.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(exponential.mean() + exponential.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(exponential.mean() - exponential.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram
plt.show()
```

```python
# "Additionally" Exercise:


country_homes = np.random.normal(5, .5, 100)
city_homes = np.random.normal(10, 1, 100)
all_homes = country_homes + city_homes

# Plot a histogram.
plt.hist(all_homes)

# Add a vertical line at the mean.
plt.axvline(all_homes.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(all_homes.mean() + all_homes.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(all_homes.mean() - all_homes.std(), color='b', linestyle='dashed', linewidth=2)

# Print the histogram.
plt.show()
```
